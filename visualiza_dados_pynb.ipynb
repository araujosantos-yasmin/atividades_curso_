```python
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
import random

dados1 = random.sample(range(100), k=20)
dados2 = random.sample(range(100), k=20)

plt.plot(dados1, dados2) # pyplot gerencia a figura e o eixo
```

*Figura com eixo como varivel*

```python
import numpy as np

x = range(5)
x = np.array(x) # temos que converter para um array numpy, seno o plot no consegue fazer operaes.

fig, ax = plt.subplots(1, 2, figsize=(12, 5)) # Cria figura com subplots: 1 linha, 2 colunas e eixos

print("Tipo de ax = ", type(ax))
print("Contedo de ax[0] = ", ax[0])
print("Contedo de ax[1] = ", ax[1])

ax[0].plot(x, x, label='eq_1') # cria grfico sobre eixo 0
ax[0].plot(x, x**2, label='eq_2') # cria grfico sobre eixo 0
ax[0].plot(x, x**3, label='eq_3') # cria grfico sobre eixo 0
ax[0].set_xlabel('Eixo x')
ax[0].set_ylabel('Eixo y')
ax[0].set_title("Grfico 1")
ax[0].legend()

ax[1].plot(x, x, 'r--', label='eq_1') # cria grfico sobre eixo 1
ax[1].plot(x**2, x, 'b--', label='eq_2') # cria grfico sobre eixo 1
ax[1].plot(x**3, x, 'g--', label='eq_3') # cria grfico sobre eixo 1
ax[1].set_xlabel('Novo Eixo x')
ax[1].set_ylabel('Novo Eixo y')
ax[1].set_title("Grfico 2")
ax[1].legend()
```

*Figura sem eixo como varivel*

```python
x = range(5)
x = np.array(x) # temos que converter para um array numpy, seno o plot no consegue fazer operaes.

fig = plt.subplots(figsize=(12, 5)) # Cria figura sem eixo
plt.subplot(121) # Adiciona um grid de subplots a figura: 1 linha, 2 colunas - Figura 1
plt.plot(x, x, label='eq_1')
plt.plot(x, x**2, label='eq_2')
plt.plot(x, x**3, label='eq_3')
plt.title("Grfico 1")
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.legend()

plt.subplot(122) # Adiciona um grid de subplots a figura: 1 linha, 2 colunas - Figura 2
plt.plot(x, x, 'r--', label='eq_1')
plt.plot(x**2, x, 'b--', label='eq_2')
plt.plot(x**3, x, 'g--', label='eq_3')
plt.title("Grfico 2")
plt.xlabel('Novo eixo x')
plt.ylabel('Novo eixo y')
plt.legend()
```

Biblioteca Seaborn

```python
import seaborn as sns

# Configurando o visual do grfico. Leia mais em https://seaborn.pydata.org/generated/seaborn.set.html#seaborn.set
sns.set(style="whitegrid") # opes: darkgrid, whitegrid, dark, white, ticks

df_tips = sns.load_dataset('tips')

print(df_tips.info())
df_tips.head()
```

Funo Barplot

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

sns.barplot(data=df_tips, x='sex', y='total_bill', ax=ax[0])
sns.barplot(data=df_tips, x='sex', y='total_bill', ax=ax[1], estimator=sum)
sns.barplot(data=df_tips, x='sex', y='total_bill', ax=ax[2], estimator=len)
```

```python
plt.figure(figsize=(10, 5))

ax = sns.barplot(x="day", y="total_bill", data=df_tips)

ax.axes.set_title("Venda mdia diria", fontsize=14)
ax.set_xlabel("Dia", fontsize=14)
ax.set_ylabel("Venda mdia ", fontsize=14)
ax.tick_params(labelsize=14)
```

Funo CountPlot()

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df_tips, x="day")
```

Funo Scartterplot()

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df_tips, x="total_bill", y="tip")
```
