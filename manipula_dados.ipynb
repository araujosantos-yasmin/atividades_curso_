ETAPA DE CAPTURA E TRANSFORMAÇÃO/ PADRONIZAÇÃO DOS DADOS

```python
df_selic = pd.read_json("https://api.bcb.gov.br/dados/serie/bcdata.sgs.11/dados?formato=json") #Armazenameno da leitura do arquivo JSON em uma variável

print(df_selic.info()) # Print das informações da variável pela função '.info()'
```

REMOVER LINHAS DUPLICADAS

```python
df_selic.drop_duplicates(keep='last', inplace=True)
# 1. Método '.drop_duplicates' para remover linhas duplicadas
# 2. Parâmetro 'keep = 'last'' para manter a última ocorrência da linha duplicada
# 3. Parâmetro 'inplace = True' para a alteração ser salva diretamente no Dataframe
```

CRIAR NOVAS COLUNAS

```python
from datetime import date
from datetime import datetime as dt

data_extracao = date.today() # Variável armazena o método que fornece a data de extração das informações

df_selic['data_extracao'] = data_extracao # A data das extrações é adicionada a uma coluna
df_selic['responsavel'] = "Autora"

print(df_selic.info())
df_selic.head()
```

MÉTODO TO_DATETIME() E ASTYPE()

```python
df_selic['data'] = pd.to_datetime(df_selic['data'], dayfirst=True) # Coluna 'data' foi convertida para formato Datetime pelo método '.to_datetime' # Começa com o dia
df_selic['data_extracao'] = df_selic['data_extracao'].astype('datetime64[ns]') # Coluna 'data_extração' já está em formato Datetime. # '.astyp
e' formata em convenção (ano primeiro)

print(df_selic.info())
df_selic.head()
```

SERIES.STR

```python
df_selic['responsavel'] = df_selic['responsavel'].str.upper()

df_selic.head()

 #ao selecionar uma coluna, que é uma serie, se aplica o método .upper() a todos os seus objetos
 # todos se tratam de uma string, que é especificada: '.str.upper()' e o método é passado a todos os valores
```

MÉTODO SORT_VALUES()

```python
df_selic.sort_values(by='data', ascending=False, inplace=True)

df_selic.head()

# organizando valores da coluna data em ordem decrescente e atualizando diretamente no Dataframe
```

MÉTODO RESET_INDEX() E SET_INDEX()

```python
df_selic.reset_index(drop= True, inplace=True) #redefine o indice usando o padrão, sem manter a coluna anterior.

df_selic.head()
```

```python
lista_novo_indice = [f'selic_{indice}' for indice in df_selic.index]

print(lista_novo_indice[:5])
```

```python
df_selic.set_index(keys=[lista_novo_indice], inplace=True) #utilizando método '.set_index()' para criar novos indices a partir da lista 'lista_novo_indice' como key

df_selic.head()
```

```python
print(df_selic['valor'].idxmin())
print(df_selic['valor'].idxmax())
```

FILTROS COM TESTES BOOLEANOS

```python
teste = df_selic['valor'] < 0.01 # armazena os valores que satisfazem a condição em uma variável

print(type(teste))
teste[:5] # slice dos valores até a quinta posição
```

```python
teste2 = df_selic['data'] >= pd.to_datetime('2020-01-01')
# teste lógico para datas maiores ou igual a '2020 - 01 - 01'
# a conversão 'pd.to_datetime' precisa ser feita, para a data não ser interpretada com string

print(type(teste2))
teste2[:5]
```

```python
teste3 = (df_selic['valor'] < 0.01) & (df_selic['data'] >= pd.to_datetime('2020-01-01'))

teste4 = (df_selic['valor'] < 0.01) | (df_selic['data'] >= pd.to_datetime('2020-01-01'))

# operadores lógicos na seleção dos valores que são armazenados em variáveis

print("Resultado do AND:\n")
print(teste3[:3])

print("Resultado do OR:\n")
print(teste4[:3])
```

*Filtro loc com condicional*

```python
filtro1 = df_selic['valor'] < 0.01

df_selic.loc[filtro1]
```

```python
# filtro loc com considicional e operadores lógicos
data1 = pd.to_datetime('2020-01-01')
data2 = pd.to_datetime('2020-01-31')

filtro_janeiro_2020 = (df_selic['data'] >= data1) & (df_selic['data'] <= data2)

df_janeiro_2020 = df_selic.loc[filtro_janeiro_2020] # filtro aplicado para valores de janeiro de 2020 armazenados em um DF
df_janeiro_2020.head()
```

```python
data1 = pd.to_datetime('2019-01-01')
data2 = pd.to_datetime('2019-01-31')

filtro_janeiro_2019 = (df_selic['data'] >= data1) & (df_selic['data'] <= data2)

df_janeiro_2019 = df_selic.loc[filtro_janeiro_2019]
df_janeiro_2019.head()
```

```python
print('Mínimo geral = ', df_selic['valor'].min())
print('Mínimo janeiro de 2019 = ', df_janeiro_2019['valor'].min())
print('Mínimo janeiro de 2020 = ',df_janeiro_2020['valor'].min(), '\n')

print('Máximo geral = ', df_selic['valor'].max())
print('Máximo janeiro de 2019 = ', df_janeiro_2019['valor'].max())
print('Máximo janeiro de 2020 = ',df_janeiro_2020['valor'].max(), '\n')

print('Média geral = ', df_selic['valor'].mean())
print('Média janeiro de 2019 = ', df_janeiro_2019['valor'].mean())
print('Média janeiro de 2020 = ',df_janeiro_2020['valor'].mean(), '\n')
```
